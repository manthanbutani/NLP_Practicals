```python
import requests   
from bs4 import BeautifulSoup as bs 
import re 
import nltk
import matplotlib.pyplot as plt
from wordcloud import WordCloud
import os
```

```python
hp_reviews=[]
for i in range(1,30):
    hp=[]  
    url="https://www.flipkart.com/hp-omen-ryzen-5-hexa-core-4600h-8-gb-512-gb-ssd-windows-10-home-4-graphics-nvidia-geforce-gtx-1650-ti-15-en0001ax-gaming-laptop/product-reviews/itmf28adfae66c40?pid=COMFSTFFPEWNUCH5&lid=LSTCOMFSTFFPEWNUCH5DUBIX4&marketplace=FLIPKART"+str(i)
    response = requests.get(url)
    soup = bs(response.content,"html.parser")# creating soup object to iterate over the extracted content 
    reviews = soup.findAll("div",attrs={"class","t-ZTKy"})# Extracting the content under specific tags  
    for i in range(len(reviews)):
        hp.append(reviews[i].text)  
        hp_reviews+=hp
```

```python
import json
json_file = json.dumps(hp_reviews)
json_file
```

```python
with open("Flipkart_reviews.json", "w") as outfile:
    outfile.write(json_file)
```

```python
import pandas as pd
df = pd.read_json("Flipkart_reviews.json")
df.to_excel (r'Flipkart_reviews.xlsx', index = None, header=True)
df
```

```python
import nltk
nltk.download('wordnet')
```

```python
df = df.replace('\n','', regex=True)
df.fillna("Not Available")
df.rename(columns={0: 'Reviews'})
```

```python
from nltk.tokenize import word_tokenize, RegexpTokenizer
```

```python
words = []
for i in df[0]:
    tokenizer = RegexpTokenizer(r'\w+')
    word = tokenizer.tokenize(i)
    words.extend(word)
words
```

```python
len(words)
```

```python
from nltk import FreqDist
frequency = FreqDist(words)
frequency
```

```python
unique_words = []
for i in frequency:
    unique_words.append(i)
len(unique_words)
```

```python
from nltk import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
for i in unique_words:
    print("{} : {}".format(i, lemmatizer.lemmatize(i)))
    lem_word = [] 
    for i in words: 
        lem_word.append(lemmatizer.lemmatize(i))
    lem = [lem_word] 
        
```

```python
from nltk import bigrams, trigrams
from collections import Counter, defaultdict
```

```python
uni = {w1: 0 for i in lem for w1 in i  }
```

```python
bigs = list(bigrams(lem_word))
bigs[:10]
```

```python
bi = defaultdict(lambda: defaultdict(lambda: 0))
```

```python
for sentence in lem:
    for w1, w2 in bigrams(sentence, pad_right=True, pad_left=True):
        bi[(w1)][w2] += 1
```

```python
for w1 in bi:
    total_count = float(sum(bi[w1].values()))
    for w2 in bi[w1]:
        bi[w1][w2] /= total_count
```

```python
dict(bi["laptop"])
```

```python
trigs = list(trigrams(lem_word))
trigs[:20]
```

```python
tri = defaultdict(lambda: defaultdict(lambda: 0))
```

```python
for sentence in lem:
    for w1, w2, w3 in trigrams(sentence, pad_right=True, pad_left=True):
        tri[(w1, w2)][w3] += 1
```

```python
for w1_w2 in tri:
    total_count = float(sum(tri[w1_w2].values()))
    for w3 in tri[w1_w2]:
        tri[w1_w2][w3] /= total_count
```
